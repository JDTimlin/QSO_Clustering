### Code for plotting multiple correlation functions at once

High-$z$ clustering constraints are illustrated in Figure~6, where we
show the quasar ``bias'' (the observed over-density of quasars as
compared to what is theoretically expected) as a function of redshift.
Three different models are shown.  The model with the largest bias at
high redshift assumes the ``status quo'' at low redshift has persisted
from the same mode at early times: quasars simply track the merger
histories of their host dark matter halos. Note that, because the two
high-$z$ data points from \citet*{sso+07} only probe the bright-end of
the QLF (since they lie to the left of the solid line marked $i <
20.2$), they only probe {\em merger histories} not {\em accretion
  histories}.  They would therefore {\em agree with the merger-history
  model regardless of the underlying cosmological truth}.

To constrain the physics of quasar feedback at early times, one needs
to push clustering measurements fainter than the current high-$z$
spectroscopic limit of $i=20.2$. Figure~6 demonstrates that, as the
survey limit is pushed fainter, ``feedback'' accretion modes (dotted
and dot-dash lines to the left of the solid line marked $i < 23.3$)
become visible.  The dotted line represents uniform ``long-lived''
growth of quasars from early times to $z=2.2$, which can be probed and
distinguished from the merger-history model up to redshifts above
$z=5$.  If this is true, then a fainter quasar at high redshift would
demonstrate that bright quasars are far more clustered than faint
quasars. This extra information would be revealed because a deeper
survey could ``see'' the faint-end of the quasar luminosity function
to higher redshifts and thus distinguish the cosmological {\em
  accretion history} of quasars from their {\em merger histories}.


At $i\sim23$ it is possible to distinguish very strong feedback, in
which quasars are strongly accreting and short-lived (the dot-dashed
line) from the other models, out to about $z\sim4$.  This can be
revealed because sampling the faint-end of the quasar luminosity
function allows us to see all quasars accreting in all modes. The
brightest quasars are both in the most dense (and thus most gas-rich)
environments and are all accreting at Eddington---so they merely tell
us that they ``turned on'' when a halo of a certain mass channeled gas
to their central engine.  As this figure can be difficult to interpret
for someone that has not worked on clustering, we simply stress that
{\em with the aid of multiwavelength data in Stripe 82 it is possible
  to push high-$z$ clustering measurements faint enough to break the
  degeneracy between the discussed models, thereby enabling powerful
  conclusions about the influence of quasar feedback on the evolution
  of galaxies in the early Universe.}

```python
## %matplotlib inline
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import interpolate
from scipy.optimize import curve_fit
```

```python
### SPLITTING BY IMAG/Z
alli = '../Compute_correlation/SpSh_angcor_RAcut_zcut_Agcut_2classification.txt'
lowi = '../Compute_correlation/imag_split/SpSh_angcor_lowimag_new.txt'
highi = '../Compute_correlation/imag_split/SpSh_angcor_highimag_new.txt'
midi = '../Compute_correlation/imag_split/SpSh_angcor_202imag.txt'

alliJK = '../Compute_correlation/SpSh_angcor_RAcut_zcut_Agcut_2classification_JK.txt'
lowiJK = '../Compute_correlation/imag_split/SpSh_angcor_lowimag_new_JK.txt'
highiJK ='../Compute_correlation/imag_split/SpSh_angcor_highimag_new_JK.txt'
midiJK = '../Compute_correlation/imag_split/SpSh_angcor_202imag_JK.txt'

datfiles = [alli,lowi,highi,midi]
JKfiles = [alliJK,lowiJK,highiJK,midiJK]
```

```python
#Loop over the data files list to open and plot the correlation function
separation = []
wnames = []
RRnames = []
for infile in datfiles:
    fileopen = open(infile,'rw')
    header = fileopen.readline()
    data = fileopen.readlines() 

    th = []
    w = []
    RR = []

    for i in range(len(data)):
        t = float(data[i].split()[0]) #Theta
        x = float(data[i].split()[4]) #w(theta) 
        rr = float(data[i].split()[3]) #RR 
        w.append(x)
        RR.append(rr)
        th.append(t)
    #Put the w and RR values into an array to call for Jackknife calculation
    wnames.append(w)
    RRnames.append(RR)
    separation.append(th)
```

```python

sigma = []
for h in range(len(JKfiles)):
    size = float(len(separation[h]))
    jkdat = open(JKfiles[h],'rw')
    #Read the header of the jackknife file and assign the number of jackknife resamples done for the sample
    jkhead = jkdat.readline()
    print jkhead
    jknum= np.float(jkhead.split()[7])
    #Pull out the proper info from the file (RR,Xi,etc.) for each jackknife and put into array
    RRjk = np.zeros([jknum,int(size)])
    Xijk = np.zeros([jknum,int(size)])
    Thjk = np.zeros([jknum,int(size)])
    
    num=jkdat.readlines()
    row = 0
    k=0

    for j in range(len(num)):
        #At the end of a Jackknife run, I create a new line called 'STEP_DONE' which tells me that my next JK routine is running and separates the two.
        if num[j].split()[0] == 'STEP_DONE':
            row +=1
            k=0
        #For each of the JK runs, append the info into arrays
        else:
            Thjk[row,k] = float(num[j].split()[0])
            RRjk[row,k] = float(num[j].split()[3])
            Xijk[row,k] = float(num[j].split()[4])
            k+=1
    #Compute the main diagonal of the covariance matrix (Update to include full covariance matrix (double loop?))
    c=0
    C=[]
    for i in range(len(Thjk)):
        sig = (np.sqrt(np.asarray(RRjk[i])/np.asarray(RRnames[h]))*(Xijk[i]-np.asarray(wnames[h])))**2
        c+=sig
        if i+1 == len(Xijk):
            C.append(c)
            sigma.append(C[0])
print sigma
```

```python
#Following Limber was computed with cosmology (H0,oM,oL)=(70 0.274489795918 0.725510204082)

##This mclimber was done for allz 2.9<2<5.15
mclimber_all = [0.0018901069830166064, 0.0015469541773802007, 0.0012618003172005974, 0.0010646848849014826, 0.00092659609065711736, 0.0008121011076479759, 0.00069717733505905916, 0.00057620099028505515, 0.00045441356648772189, 0.0003334100976949317, 0.00022407216949898401, 0.00013460720655439581, 6.8890660458582111e-05, 2.8945118539565589e-05, 8.5993807832771192e-06, 2.6124134574857943e-06, -2.353262803259482e-06, -1.0507113239217212e-06, -4.7148295416947918e-07, -9.8152287015064401e-08]
limber = np.asarray(mclimber_all)*np.pi / 0.7**3 /0.94

#Interpolate Limber (you get the same function back) to plug in any z in the range (as opposed to set z values)
Limber = interpolate.interp1d(np.logspace(-1.3,2.5,20),limber)
```

```python
#Fit Limber to wtheta to get ideal bias value
def L(theta,b):
    l = b**2 * Limber(theta)
    return l

gdx = (np.asarray(separation[0])>4) & (np.asarray(separation[0])<48)


allibias, allicov = curve_fit(L, np.asarray(separation[0])[gdx], np.asarray(wnames[0])[gdx], sigma = sigma[0][gdx]**0.5,absolute_sigma=True)
lowibias, lowicov = curve_fit(L, np.asarray(separation[1])[gdx], np.asarray(wnames[1])[gdx], sigma = sigma[1][gdx]**0.5,absolute_sigma=True)
highibias, highibcov = curve_fit(L, np.asarray(separation[2])[gdx], np.asarray(wnames[2])[gdx], sigma = sigma[2][gdx]**0.5,absolute_sigma=True)
midibias, midibcov = curve_fit(L, np.asarray(separation[3])[gdx], np.asarray(wnames[3])[gdx], sigma = sigma[3][gdx]**0.5,absolute_sigma=True)

print allibias
print lowibias
print highibias
print midibias
```
